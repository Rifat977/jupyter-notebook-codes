```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, r2_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv("titanic.csv")
df
```

```python
df.isnull().sum()
```

```python
# df2 = df.fillna(0)
# df2 = df.fillna(method='bfill')
# df2 = df.fillna(method='pad')
# df.dropna(how='all')
# df2 = df['Age'].fillna(value=df['Age'].mean())
# df['Age'] = df2
# df = df.drop('Cabin', axis=1)
# df.isnull().sum()
df2 = df['Age'].fillna(value=df['Age'].mean())
df['Age'] = df2
df = df.drop('Cabin', axis=1)
df = df.fillna(method='bfill')
df.isnull().sum()
```

```python
encoder = LabelEncoder()
df['Sex'] = encoder.fit_transform(df['Sex'])
df['Ticket'] = encoder.fit_transform(df['Ticket'])
df['Embarked'] = encoder.fit_transform(df['Embarked'])
df
```

```python
features = df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']]
features
```

```python
target = df['Survived']
target
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.3, random_state=6)
```

```python
mdl = LinearRegression()
mdl.fit(x_train, y_train)
```

```python
predict = mdl.predict(x_test)
r2_score(y_test, predict) * 100
```

```python
lmodel = LogisticRegression(max_iter=1000)
lmodel.fit(x_train, y_train)
```

```python
lpredict = lmodel.predict(x_test)
accuracy_score(y_test, lpredict) * 100
```
